```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
#Opgave 51
titanic.fare.plot(kind = 'hist', title = 'histogram voor fare', bins = 20)
```

```python
#Opgave 52
import matplotlib.pyplot as plt
titanic.fare.plot(kind = 'kde', title = 'dichtheidsfunctie voor fare') # hier wordt onderliggend van een norm verdeling uitgegaan
plt.xlabel("fare")
plt.ylabel("dichtheid")
ax = plt.gca() #Get the Current Axes
ax.set_xlim(-30, 250)
```

```python
titanic.fare.plot(kind = 'density')
plt.xlabel("fare")
plt.ylabel("dichtheid")
ax = plt.gca() #Get the Current Axes
ax.set_xlim(-30, 250)
```

```python
#Opgave 54: histogram voor variabele leeftijd
titanic.age.plot(kind = 'hist', title = 'histogram voor leeftijd', bins = 15)
```

```python
#Opgave 55: dichtheidsfunctie voor de variabele 'age'
titanic.age.plot(kind = 'kde', title = 'dichtheidsfunctie voor age')
```

```python
#Opgave 56: skewness parameter voor 'age
titanic.age.skew()
```

```python
#Opgave 57
print(titanic.age.mean())
print(titanic.age.std())
```

```python
import scipy.stats as stats
```

```python
#Opgave 58
stats.norm.cdf(30, titanic.age.mean(), titanic.age.std()) - stats.norm.cdf(20, titanic.age.mean(), titanic.age.std())
```

```python
#Opgave 59: TEST opgave 58 op steekproefgegevens
titanic_leeftijd20_leeftijd30 = titanic[(titanic.age >= 20) & (titanic.age <= 30)]
print(len(titanic_leeftijd20_leeftijd30))
```

```python
print(len(titanic_leeftijd20_leeftijd30) / len(titanic))
```

```python
#Vraag 60: om af te sluiten...
import matplotlib.pyplot as plt
```

```python
#a: meerdere subplots tegelijk in python
f, ax_array = plt.subplots(3, 2, figsize=(20,10)) # 3 rijen en 2 kolommen

ax_array[0,0].hist(titanic.fare, bins = 20, color = 'c')
ax_array[0,0].set_title('Histogram van fare')
ax_array[0,0].set_xlabel('categorien')
ax_array[0,0].set_ylabel('aantal')

ax_array[0,1].hist(titanic.age, bins = 20, color = 'tomato')
ax_array[0,1].set_title('Histogram van leeftijd')
ax_array[0,1].set_xlabel('categorien')
ax_array[0,1].set_ylabel('aantal')

ax_array[1,0].boxplot(titanic.fare.values)
ax_array[1,0].set_title('Boxplot van fare')
ax_array[1,0].set_xlabel('fare')
ax_array[1,0].set_ylabel('prijs')

ax_array[1,1].boxplot(titanic.age.values)
ax_array[1,1].set_title('Boxplot van age')
ax_array[1,1].set_xlabel('age')
ax_array[1,1].set_ylabel('leeftijd')


ax_array[2,0].scatter(titanic.age, titanic.fare, alpha=0.15)
ax_array[2,0].set_title('Scatterplot van age tov fare')
ax_array[2,0].set_xlabel('age')
ax_array[2,0].set_ylabel('fare')

#afstand tussen plots 
plt.tight_layout() 
#lege subplot onderaan weg (3 bij 2 = 6 plaatsen)
ax_array[2,1].axis('off')

plt.show()
```

```python
#b: heatmap: enkel numerische variabelen worden opgenomen
sns.heatmap(titanic.corr(), annot=True, cmap= "coolwarm") #annot = True: cijfers erbij; cmap: kleuroptie
plt.show()
```

```python
#c: voorwaardelijke kans
fig, axs = plt.subplots(1, 2, figsize=(10,4))
for i, sex in enumerate(['female', 'male']):
    p = titanic[titanic['sex'] == sex]['survived'].value_counts(normalize=True)
    p.plot(kind='bar', ax=axs[i])
    axs[i].set_title('Histogram overlevend - {:0.1%} Survived ({})'.format(p[1], sex))
    axs[i].set_ylabel('% of passagiers')
    axs[i].set_xlabel('0 = dood, 1 = overleefd')
```
